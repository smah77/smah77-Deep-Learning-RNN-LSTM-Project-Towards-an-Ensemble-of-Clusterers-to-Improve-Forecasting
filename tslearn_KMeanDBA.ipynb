```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tslearn.utils import to_time_series_dataset, save_timeseries_txt
from tslearn.clustering import TimeSeriesKMeans
from tslearn.preprocessing import TimeSeriesScalerMeanVariance, TimeSeriesResampler

data = "D:/DKE STUDIUM/4th Semester/DBSE Team project/train_1.csv/train_1_50ts.csv"
train = pd.read_csv(data, nrows=100)
train.head()
#train.drop('Page', axis=1, inplace=True)
#data1 = np.genfromtxt(train, dtype='unicode', delimiter=',', names=None) 
```

```python
from tslearn.clustering import TimeSeriesKMeans
import tslearn.utils

train.fillna(value=0, inplace=True) #replace all the NaN values with Zero's in a column of a pandas dataframe
ts = to_time_series_dataset(train.as_matrix(columns=train.columns[520:]))
save_timeseries_txt("C:/Users/Dell/Desktop/ts_test2.txt", ts)
km = TimeSeriesKMeans(n_clusters=5, metric="dtw", max_iter=5, max_iter_barycenter=5, verbose=False, random_state=0).fit(ts)
#print(ts)
#print('Cluster centroids: ')
#print(km.cluster_centers_)
save_timeseries_txt("C:/Users/Dell/Desktop/ts_km.txt", km.cluster_centers_)
```
